```python
df = pd.read_csv('data/text_emotion.csv')
df.head()
```

```python
import re

# use of re.compile makes the regex opr. faster if the same regex is applied multiple times, 
# eg. here its applied for each tweet in the df['content'] column
# regular exp. char class which is [@<any alphanumeric string>] will match with any such reference in the content
regex_user_reference = re.compile('@[a-zA-Z0-9]+') 
# references = []
for entry in df['content']:
    # references.append(regex_user_reference.findall(entry))
    regex_user_reference.findall(entry)
df['content'] = df['content'].str.replace(regex_user_reference,'<user>', regex=True)
```

```python
regex_websites = re.compile('http://[www.]*[a-zA-Z0-9]+.[a-z]+/[a-zA-Z0-9//]*')
# websites = []
for entry in df['content']:
    regex_websites.findall(entry)
df['content'] = df['content'].str.replace(regex_websites,'<url>', regex=True)
```

```python
# df['content'].head(10)
df.head(10)
```

```python
tweets = df['content'].copy(deep=True)
# to take out chars like ... or :
regex_extrachars = re.compile('[\.\:]+')
df['content'] = df['content'].str.replace(regex_extrachars,' ', regex=True)
```

```python
df.head(10)
```

```python
list_split_tweets = []
counter = 0
for entry in df['content']:
    if counter == 10:
        break   
    print(entry)    
    entry = entry.translate({ ord(c): " "+c+" " for c in "!@#$%^&*()[]{}';:,./?\|`~-=_+" }) 
    split_tweet = entry
    print(split_tweet)
    counter += 1
    list_split_tweets.append(split_tweet.split(" "))   
```

```python
WordToVec = dict()
with open('data/glove.6B.50d.txt', encoding='utf-8') as file:
    for line in file:
        entry_list = line.split(" ")
        word = entry_list[0] 
        word_vector = [float(val) for val in entry_list[1:]]
        WordToVec[word] = word_vector
```

```python
len(WordToVec['the'])
```

```python
list_sentences_vectors = []
for split_tweet in list_split_tweets:
    sentence = []
    for word in split_tweet:
        try:    
            sentence.append(WordToVec[word.lower()])
        except Exception as e:
            print("This word is not present in the embeddings:", e)
        senToVector = np.array(sentence)
    list_sentences_vectors.append(senToVector)
```
